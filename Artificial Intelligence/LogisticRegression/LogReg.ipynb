```python
import pandas as pd
data = pd.read_excel('reuters.xlsx')
X, y = data.text, data.topic

import nltk
nltk.download('wordnet')
nltk.download('stopwords')

lem_texts = []
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
for t in X:
    lem = [wnl.lemmatize(word) for word in str(t).split()]
    lem_texts.append(' '.join(lem))

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=500, min_df=10, max_df=0.8,
                     stop_words=nltk.corpus.stopwords.words('english'))
#max_features - к-сть слів, які використовуються для класифікації
#min_df - мін к-сть текстів, у яких міститься слово
#max_df - макс відсоток файлів, у яких міститься слово
#stop_words - шумові слова
X = cv.fit_transform(lem_texts).toarray()

from sklearn.feature_extraction.text import TfidfTransformer
tfidfconverter = TfidfTransformer()
X = tfidfconverter.fit_transform(X).toarray()

import numpy as np
labels = list(set(y))
y = np.array([labels.index(x) for x in y])

print(f'Topics: {len(set(y))}, X_shape: {X.shape}')
```

```python
import LogisticRegression
# model = LogisticRegression.LogisticRegression(X, y, 1.0, 100, regularized=True, normalization = 'l1')
model = LogisticRegression.LogisticRegression(X, y, 1.0, 50, regularized=True, normalization = 'l2')
learntParameters, final_costs = model.train(X, y, np.unique(y))
print('Number of classes', len(np.unique(y)))
#print('learntParameters(one per class): ',learntParameters)
#print('final_costs: ', final_costs)
#print(len(learntParameters))
y_pred = []
for eachData in X:
	y_pred.append(model.classify(eachData, learntParameters))
y_pred = np.array(y_pred)

print('Accuracy on training data: ',(np.sum(y_pred == y)/len(y))*100,'%')
```
